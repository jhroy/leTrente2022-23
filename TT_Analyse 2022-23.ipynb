# Analyse textuelle des publications TikTok des m√©dias qu√©b√©cois (2022-23)

Cette analyse porte sur les plus r√©centes publications TikTok des m√©dias du Qu√©bec (entre le 1<sup>er</sup> janvier 2022 et le 31 mai 2023).

Les diff√©rentes √©tapes ci-dessous correspondent √† ce qui a √©t√© effectu√© sur le corpus de publications Facebook et qui sont d√©crites dans [cet autre carnet](https://github.com/jhroy/leTrente2022-23/blob/main/FB_Analyse%202022-23.ipynb)

```python
import pandas as pan
import spacy, csv
from collections import Counter
tal = spacy.load("fr_core_news_sm")
```

```python
tt = pan.read_csv("TT_22_23.csv")
```

```python
motsNot = ["the", "in", "to", "and", "of", "|", "for", "s", "t", "full", "story", "tap", "click", "by", "with",
               "link", "more", "at", "our", "follow", "us", "don'", "forget", "website", "like"]
```

```python
for mois in tt.mois.unique():
    
    f = open("TT_22_23.csv")
    videos = csv.reader(f)
    next(videos)
    
    texte = ""
    for video in videos:
        if video[-2] == mois:
            texte = texte + video[2] + " " + video[7] + " "
            
    texte = texte.replace("\n", " ").replace("‚Äô", "'").replace("\xa0", " ").replace("\u2060", " ").replace("\u2009", " ").replace("##", "#").replace("#", " #")
    
    while "  " in texte:
        texte = texte.replace("  "," ").strip()
#     print(mois,len(texte))
    
    doc = tal(texte)
#     print(len(doc))
    
    lemmes = [token.lemma_.lower() for token in doc if token.is_stop == False and token.is_punct == False]
    lemmes = [token.lemma_.lower() for token in doc if token.is_stop == False and token.is_punct == False and token.lemma_.lower() not in motsNot and len(token.text) > 1 and "/" not in token.text and "@" not in token.text]
#     print(mois,len(lemmes))
    
    freq = Counter(lemmes)
    print(freq.most_common(5))
    
    for n in freq.most_common():
#         print([jour[:7],n[0],n[1]])
        daft = open("TT_motsSeuls.csv", "a")
        punk = csv.writer(daft)
        punk.writerow([mois,n[0],n[1]])
    
    bigrammes = []
    
    for index, lemme in enumerate(lemmes[:-1]):
        bigrammes.append("{} {}".format(lemmes[index], lemmes[index+1]))
    
    freq = Counter(bigrammes)
    print(freq.most_common(5))
    
    for n in freq.most_common():
#         print([jour[:7],n[0],n[1]])
        daft = open("TT_bigrammes.csv", "a")
        punk = csv.writer(daft)
        punk.writerow([mois,n[0],n[1]])
    
    emojis = []

    for token in doc:
        if token.is_stop == False and token.is_punct == False and len(token.text) == 1:
            emojis.append(token.text)
    
    freq = Counter(emojis)
    print(freq.most_common(10))
    
    for n in freq.most_common():
#         print([jour[:7],n[0],n[1]])
        daft = open("TT_emojis.csv", "a")
        punk = csv.writer(daft)
        punk.writerow([mois,n[0],n[1]])
    
    hashtags = []

    for token in texte.split():
        if token[0] == "#":
            hashtags.append(token)
    
    freq = Counter(hashtags)
    print(freq.most_common(10))
    
    for n in freq.most_common():
#         print([jour[:7],n[0],n[1]])
        daft = open("TT_hashtags.csv", "a")
        punk = csv.writer(daft)
        punk.writerow([mois,n[0],n[1]])
```
