# Analyse des *posts* Facebook de 309 médias du Québec (2011-2023)
On commence par importer les modules nécessaires

```python
import glob, os
import pandas as pan
```

On importe tous les fichiers extraits de [CrowdTangle](https://www.crowdtangle.com/).

Le nom de chacun a la même structure.

`Historical-Report-MediasQuebec-2013-07-08-09-10-11-12.csv` signifie que c'est un rapport historique d'une liste que j'ai créée dans CrowdTangle et que j'ai baptisée `MediasQuebec`. Ici, le fichier couvre les mois 07 (juillet) à 12 (décembre) de l'année 2013.

Quand on fait une exportation historique, chaque fichier doit contenir moins de 100k publications, d'où le grand nombre de fichiers.

```python
tousfich = glob.glob(os.path.join("", "Historical*.csv"))
print(tousfich)
```

Nous avons importé 60 fichiers en tout.

```python
len(tousfich)
```

Les deux lignes de code ci-dessous vont d'abord lire tous les fichiers et les placer dans la variable `chq_fich`, puis les concaténer avec la méthode `.concat` et placer le tableau de données (*dataframe*) résultant dans la variable `fich`.

```python
chq_fich = (pan.read_csv(f, low_memory=False) for f in tousfich)
fich = pan.concat(chq_fich, ignore_index=True)
```

En faisant s'afficher le tableau `fich`, on constate qu'il contient plus de 4,5 millions d'enregistrements. Chacun est une publication d'un média dans Facebook depuis le 1er janvier 2011 jusqu'au 31 mai 2023.

```python
fich
```

Pour faciliter l'analyse, on crée deux nouvelles colonnes pour indiquer à quelle année et au cours de quel mois (paire année-mois, en fait) a été publié chaque *post*.

```python
fich["année"] = fich["Post Created"].str[:4]
fich["mois"] = fich["Post Created"].str[:7]
```

Simple vérification pour voir si nos variables ont bel et bien été créées.

```python
fich
```

Combien de publications avons-nous à chaque année?

On voit que 2020 (l'année de la COVID) a été celle où les médias ont le plus mobilisé Facebook pour rejoindre les Québécoises et les Québécois. Les années 2022 et 2021 suivent de près, ce qui indique que les médias se servent encore énormément de ce réseau social.

```python
fich.année.value_counts()
```

Même vérification pour les mois.

On voit que les trois premiers mois de la pandémie sont sur le podium.

```python
fich.mois.value_counts()
```

On exporte cette répartition du nombre de *posts* par mois dans un petit fichier CSV.

```python
fich.mois.value_counts().to_csv("parMois.csv")
```

On examine le type des variables de notre tableau.

La variable `Total Interactions` est un `object`, ce qui signifie qu'elle n'est pas un nombre et qu'on ne pourra pas l'utiliser pour faire des sommes des interactions par mois ou par média.

```python
fich.dtypes
```

On va donc créer une nouvelle colonne appelée `inter` qui sera le résultat de la somme des neuf types d'interactions enregistrées par Facebook (commentaires, partages, *likes* et autres réactions).

```python
fich["inter"] = fich.Likes + fich.Comments + fich.Shares + fich.Love + fich.Wow + fich.Haha + fich.Sad + fich.Angry + fich.Care
```

Tout semble bien se passer 👇

```python
fich
```

On peut désormais faire une somme des interactions suscitées par les contenus journalistiques québécois dans Facebook à chaque mois entre janvier 2011 et mai 2023.

```python
fich.groupby(by="mois").inter.sum()
```

Résultat qu'on exporte dans un autre fichier (`interactionsParMois.csv`) qu'on combinera ultérieurement (dans LibreOffice) avec le fichier créé ci-dessus, `parMois.csv`.

```python
fich.groupby(by="mois").inter.sum().to_csv("interactionsParMois.csv")
```

Ici, je crée une liste des identifiants de chaque page médiatique (il y en a 309 au total) et je l'exporte dans le fichier `fbID.csv`. Cette liste servira à exclure les médias d'une autre analyse effectuée sur un corpus contrôle des publications Facebook ayant recueilli le plus d'interactions par mois.

```python
fich["Facebook Id"].value_counts()
```

```python
fich["Facebook Id"].value_counts().to_csv("fbID.csv")
```

Afin de réaliser une analyse plus fine, je calcule, avec le code ci-dessous et grâce à la méthode `.nunique()`, le nombre de médias qui, chaque mois, ont publié du matériel dans Facebook. Si j'ai 309 pages médiatiques dans mon échantillon, ce ne sont pas tous ces médias qui publient chaque mois.

On constate, par exemple, qu'en avril 2023, 295 pages ont publié du contenu, mais qu'en mai 2023, il n'y en avait plus que 278. Est-ce le début d'une désaffection de Facebook par les médias du Québec? Il faudra attendre l'année prochaine pour le savoir.

J'exporte ensuite le résultat dans le fichier `NbMediasParMois.csv`.

```python
fich.groupby(by="mois")["Page Name"].nunique()
```

```python
fich.groupby(by="mois")["Page Name"].nunique().to_csv("NbMediasParMois.csv")
```
